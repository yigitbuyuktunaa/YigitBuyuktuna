```python

class Library:
    def __init__(self, file_name="books.txt"):
        self.file_name = file_name
        self.file = open(self.file_name, "a+")

    def __del__(self):
        self.file.close()

    def list_books(self):
      self.file.seek(0)
      books = self.file.read().splitlines()
      if not books:
          print("There is no book in library.")
      else:
          print("Book List:")
          for book in books:
              book_info = book.split(',')
              print("Book Name: {}, Author: {}".format(book_info[0], book_info[1]))
    def list_books(self):
        self.file.seek(0)
        books = self.file.read().splitlines()
        if not books:
            print("There is no book in library.")
        else:
            print("Book List:")
            for book in books:
                book_info = book.split(',')
                print("Book Name: {}, Author: {}".format(book_info[0], book_info[1]))

    def add_book(self):
        title = input("Please enter the title of the book: ")
        author = input("Enter the author of the book: ")
        release_year = input("Enter release year of the book: ")
        num_pages = input("Enter number of pages: ")

        book_info = ','.join([title, author, release_year, num_pages]) + '\n'

        self.file.write(book_info)
        print("The book was added succesfully.")

    def remove_book(self):
        title_to_remove = input("Enter the book name that wants to be deleted: ")


        self.file.seek(0)
        books = self.file.read().splitlines()
        found = False

        updated_books = []


        for book in books:
            book_info = book.strip().split(',')
            if book_info[0] == title_to_remove:
                found = True
                print(f"'{title_to_remove}' was deleted succesfully.")
            else:
                updated_books.append(book)

        if not found:
            print(f"'{title_to_remove}' was not found.")
            return

        self.file.seek(0)
        self.file.truncate()
        self.file.write('\n'.join(updated_books))

    def search_book(self):
        search_criteria = input("Enter search criteria (title/author/release year/num pages): ").lower()
        search_keyword = input("Enter the keyword to search: ").lower()

        self.file.seek(0)
        books = self.file.read().splitlines()
        found_books = []

        for book in books:
            book_info = book.strip().split(',')
            if search_criteria == "title" and search_keyword in book_info[0].lower():
                found_books.append(book)
            elif search_criteria == "author" and search_keyword in book_info[1].lower():
                found_books.append(book)
            elif search_criteria == "release year" and search_keyword in book_info[2].lower():
                found_books.append(book)
            elif search_criteria == "num pages" and search_keyword in book_info[3].lower():
                found_books.append(book)

        if found_books:
            print("Matching Books:")
            for book in found_books:
                book_info = book.strip().split(',')
                print("Book Name: {}, Author: {}, Release Year: {}, Num Pages: {}".format(
                    book_info[0], book_info[1], book_info[2], book_info[3]))
        else:
            print("No matching books found.")
    def update_book(self):
        title_to_update = input("Enter the title of the book to update: ")
        self.file.seek(0)
        books = self.file.readlines()
        found = False

        updated_books = []

        for book in books:
            book_info = book.strip().split(',')
            if book_info[0] == title_to_update:
                found = True
                print("Enter the new information (leave empty to keep existing)")

                new_title = input("New title (leave empty to keep existing): ")
                new_author = input("New author (leave empty to keep existing): ")
                new_release_year = input("New release year (leave empty to keep existing): ")
                new_num_pages = input("New number of pages (leave empty to keep existing): ")

                if not new_title:
                    new_title = book_info[0]
                if not new_author:
                    new_author = book_info[1]
                if not new_release_year:
                    new_release_year = book_info[2]
                if not new_num_pages:
                    new_num_pages = book_info[3]

                updated_book_info = ','.join([new_title, new_author, new_release_year, new_num_pages]) + '\n'
                updated_books.append(updated_book_info)
            else:
                updated_books.append(book)

        if not found:
            print(f"'{title_to_update}' was not found.")
            return

        with open(self.file_name, "w") as file:
            file.writelines(updated_books)

lib = Library()
print("Welcome to Library Management System")
while True:
  print("""
  **************************************

  MENU
  1.List Books
  2.Add Book
  3.Remove Book
  4.Search Book
  5.Update Book
  q.Quit the system

  ***************************************
  """)
  menu_choice = input("Please enter the element:(1/2/3/4/5/q): ").lower()

  if menu_choice == "1":
    lib.list_books()

  elif menu_choice == "2":
    lib.add_book()

  elif menu_choice == "3":
    lib.remove_book()

  elif menu_choice == "4":
    lib.search_book()

  elif menu_choice == "5":
    lib.update_book()

  elif menu_choice == "q":
    print("Thanks for using the Library Management System. Have a good day.")
    break

  else:
    print("Invalid Number. Try Again.")








```
